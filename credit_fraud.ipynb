```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('fraudTrain.csv')
```

```python
df.head()
```

```python
label_encoder = LabelEncoder()
df['trans_date_trans_time'] = label_encoder.fit_transform(df['trans_date_trans_time'])
df['merchant'] = label_encoder.fit_transform(df['merchant'])
df['category'] = label_encoder.fit_transform(df['category'])
df['job'] = label_encoder.fit_transform(df['job'])
df['dob'] = label_encoder.fit_transform(df['dob'])
df['trans_num'] = label_encoder.fit_transform(df['trans_num'])
df['unix_time'] = label_encoder.fit_transform(df['unix_time'])
df['first'] = label_encoder.fit_transform(df['first'])
df['last'] = label_encoder.fit_transform(df['last'])
df['gender'] = label_encoder.fit_transform(df['gender'])
df['street'] = label_encoder.fit_transform(df['street'])
df['city'] = label_encoder.fit_transform(df['city'])
df['state'] = label_encoder.fit_transform(df['state'])
```

```python
df.head()
```

```python
legit = df[df.is_fraud == 0] 
fraud = df[df.is_fraud == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
legit_sample = legit.sample(7506)
```

```python
new_df = pd.concat([legit_sample,fraud])
```

```python
new_df['is_fraud'].value_counts()
```

```python
X = new_df.drop(columns = 'is_fraud',axis=1)
y = new_df['is_fraud']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3 , random_state = 1)
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
dtc.predict(X_test)
```

```python
dtc.score(X_test,y_test)
```
